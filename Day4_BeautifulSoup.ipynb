*import libraries

*save the url

*using the requests.get(),access the webpage

*using beautifulSoup(),access the HTML code





```python
from bs4 import BeautifulSoup as bs
```

```python
import requests as re
```

```python
c="https://en.m.wikipedia.org/wiki/List_of_tallest_buildings_in_India"
```

```python
page=re.get(c)
```

```python
code=bs(page.text,"html")
print(code)
```

```python
t=code.find("table")
print(t)
```

```python
print(t.text)
```

```python
url="https://en.m.wikipedia.org/wiki/List_of_tallest_buildings_in_India"
```

```python
page=re.get(url)
```

```python
l=bs(page.text,"html")
print(l)
```

```python
t=l.find("table")
print(t)
```

```python
h=t.find_all('th')
print(h)
```

```python
Headings=[i.text for i in h]
print(Headings)
```

```python
import pandas as pd
```

```python
mydf=pd.DataFrame(columns=Headings)
print(mydf)
```

```python
rows=t.find_all("tr")
print(rows)
```

```python
td=t.find_all("td")
print(td)
```

```python
data_act=[i.text for i in td]
```

```python
print(data_act)
```

```python
data_act[0:4]
```

```python
ind=0
st=0
stop=9
for i in data_act:
  if stop <=len(data_act):
    mydf.loc[ind]=data_act[st:stop]
    st=st+9
    stop=stop+9
    ind=ind+1
print(mydf)
```
